# UV Sheet Parser - Automated Setup

This notebook sets up the UV Sheet Parser environment on your machine.

**Run each cell in order.** ✅

## Step 1: Clone or Update Repository

````python
import os
import subprocess
import sys
from pathlib import Path

# Configuration
REPO_URL = "https://github.com/TomCharlesRousseau/uvsheet2openbis.git"
CLONE_PATH = Path("C:/uvsheet2openbis")  # Change this if desired

print(f"📁 Target directory: {CLONE_PATH}")
print()

if CLONE_PATH.exists():
    print(f"✅ Repository already exists at {CLONE_PATH}")
    os.chdir(CLONE_PATH)
    print("📥 Pulling latest changes...")
    result = subprocess.run(["git", "pull"], capture_output=True, text=True)
    print(result.stdout)
    if result.returncode != 0:
        print(f"❌ Error: {result.stderr}")
else:
    print(f"📥 Cloning repository...")
    result = subprocess.run(["git", "clone", REPO_URL, str(CLONE_PATH)], capture_output=True, text=True)
    print(result.stdout)
    if result.returncode != 0:
        print(f"❌ Error: {result.stderr}")
        raise RuntimeError("Failed to clone repository")
    os.chdir(CLONE_PATH)

print(f"\n✅ Working directory: {os.getcwd()}")
````

## Step 2: Check Python Version

````python
import sys

print(f"🐍 Python version: {sys.version}")
print(f"📍 Python executable: {sys.executable}")

# Check if Python 3.10+
if sys.version_info < (3, 10):
    print("\n⚠️  WARNING: Python 3.10+ is recommended!")
else:
    print("\n✅ Python version is compatible")
````

## Step 3: Upgrade pip and Install Dependencies

````python
import subprocess
import sys

print("📦 Upgrading pip...")
subprocess.run([sys.executable, "-m", "pip", "install", "--upgrade", "pip"], capture_output=True)
print("✅ pip upgraded")

print("\n📦 Installing requirements...")
result = subprocess.run([sys.executable, "-m", "pip", "install", "-r", "requirements.txt"], capture_output=True, text=True)
print(result.stdout)

if result.returncode != 0:
    print(f"❌ Error: {result.stderr}")
    raise RuntimeError("Failed to install dependencies")
else:
    print("✅ All dependencies installed successfully")
````

## Step 4: Verify Installation (After Kernel Restart)

````python
import importlib
import os
from pathlib import Path

# Make sure we're in the right directory
os.chdir("C:/uvsheet2openbis")

packages = ["pandas", "openpyxl", "pybis", "keyring", "streamlit"]

print("🔍 Checking installed packages:\n")
all_ok = True

for package in packages:
    try:
        mod = importlib.import_module(package)
        version = getattr(mod, "__version__", "unknown")
        print(f"✅ {package:<15} {version}")
    except ImportError:
        print(f"❌ {package:<15} NOT FOUND")
        all_ok = False

if all_ok:
    print("\n✅ All packages installed!")
else:
    print("\n⚠️  Some packages are missing. Please restart kernel and retry.")
````

## Step 5: Check Project Structure

````python
from pathlib import Path

required_files = [
    "main.py",
    "streamlit_app.py",
    "requirements.txt",
    "config/settings.json",
    "utils/person_lookup.py",
    "openbis/object_manager.py",
    "openbis/connection.py",
    "excel/excel_parser.py",
]

print("📂 Checking project structure:\n")
all_ok = True

for file in required_files:
    path = Path(file)
    if path.exists():
        print(f"✅ {file}")
    else:
        print(f"❌ {file} - MISSING")
        all_ok = False

if all_ok:
    print("\n✅ Project structure is complete!")
else:
    print("\n⚠️  Some files are missing!")
````
